```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

dataframe = pd.read_csv("/content/car data.csv")
```

```python
dataframe.head()
```

```python
dataframe.info()
```

```python
dataframe.isnull().sum()
```

```python
#encode car name column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataframe['Car_Name'] = le.fit_transform(dataframe['Car_Name'])
#enocde fuel_type columm
dataframe['Fuel_Type'] = le.fit_transform(dataframe['Fuel_Type'])
#encode seller_type column
dataframe['Seller_Type'] = le.fit_transform(dataframe['Seller_Type'])
dataframe['Transmission'] = le.fit_transform(dataframe['Transmission'])

```

```python
dataframe.head()
```

```python
#scaling techniques on dataset
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dataframe = pd.DataFrame(scaler.fit_transform(dataframe), columns=dataframe.columns)
#
```

```python
dataframe.head()
```

```python
#training the dataset
x = dataframe.drop(['Selling_Price'], axis=1)
y = dataframe['Selling_Price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
prediction = model.predict(x_test)
```

```python
r2_score = metrics.r2_score(y_test, model.predict(x_test))
```
